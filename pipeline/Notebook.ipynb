```python
# 1. Re-initialize the iterator (Crucial!)
df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)

# 2. Process everything in ONE go
is_first_chunk = True

for df_chunk in df_iter:
    if is_first_chunk:
        # Create table (replace)
        df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
        is_first_chunk = False
        print(f"Created table and inserted first chunk of {len(df_chunk)}")
    else:
        # Append data
        df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
        print(f"Inserted another chunk of {len(df_chunk)}")

print("Done! All data is now in Postgres.")
```
